```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from transformers import AutoTokenizer, AutoModelForMaskedLM
import torch
import numpy as np
```

```python
tokenizer = AutoTokenizer.from_pretrained("HooshvareLab/bert-base-parsbert-uncased")
model = AutoModelForMaskedLM.from_pretrained("HooshvareLab/bert-base-parsbert-uncased")
```

```python
model.config.output_hidden_states = True
```

```python
def embedding_system(text="" ,tokenizer=tokenizer):
    inputs = tokenizer(text, return_tensors="pt")
    with torch.no_grad():
        outputs = model(**inputs)
    # Extract the last hidden state (this contains the embeddings)
    hidden_states = outputs.hidden_states
    # Extract embeddings from the last hidden state
    last_hidden_state = hidden_states[-1]
    # Convert the embeddings to a numpy array (optional)
    word_embedding_numpy= last_hidden_state.cpu().numpy()
    return word_embedding_numpy[0,1]
```

```python
data = pd.read_csv('./../data/data.csv')
wrong_sentence_word_embedding=[]
correct_sentence_word_embedding=[]
for index in data.index:
    list_1=[]
    list_2=[]
    for token in data.at[index,"word_tokens_wrong"]:
        list_1.append(embedding_system(token))
    for token in data.at[index,"word_tokens_correct"]:
        list_2.append(embedding_system(token))
    wrong_sentence_word_embedding.append(list_1)
    correct_sentence_word_embedding.append(list_2)


```
